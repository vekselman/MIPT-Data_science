*Задание:* Создайте матрицу размера n d, где n — число предложений. Заполните ее: элемент с индексом (i, j) в этой матрице должен быть равен количеству вхождений j-го слова в i-е предложение. У вас должна получиться матрица размера 22 254.

```python
import numpy as np
import re
from __future__ import print_function
f = open('sentences.txt', 'r') # открыли файл
lines = [] # список lines будет списком списков, содержащими непустые слова с приведённым нижним регистром
for l in f:
    l = l.lower()
    l = re.split('[^a-z]',l)
    l = filter(None, l)
    lines.append(l)

word_index = dict() # словарь в котром будут содержаться уникальные слова как ключи
word_list = [] # скисок уникальных слов, выстроенных по номеру в списке
i = 0
for sentence in lines:
    for word in sentence:
        if word not in word_index:
            word_index[word] = i
            word_list.append(word)
            i = i + 1
#print (word_list)
#print (lines)

matrix = [] # хочу создать список списков (что и будет по моему мнению матрицей) из 0 и 1
for word in word_list: # пробегаю переменной word по списку слов word_list
    
    if word in lines: # если слово встречается в предложении (список lines)
        matrix[lines][word] = 1 # то записываю цифру 1 в строку с номером предложения (здесь чувствую что-то не правильно) и номером слова
    else:
        matrix[lines][word] = 0 # в противном случае, записываю цифру 0

# НО НЕ РАБОТАЕТ :-(


```

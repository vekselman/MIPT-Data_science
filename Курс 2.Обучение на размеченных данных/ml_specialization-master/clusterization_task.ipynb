```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('checkins.dat', delimiter='|', skipinitialspace=True)
data = data.dropna()
data.head()
```

```python
data.shape
```

```python
from sklearn.cluster import MeanShift
```

```python
data.info()
```

```python
clust = MeanShift(bandwidth=0.1, n_jobs=4)
clust.fit(data.values[:100000, 3:5])
labels = clust.labels_
cluster_centers = clust.cluster_centers_
```

```python
print cluster_centers
```

```python
filtered_centers = []
for label in np.unique(labels):
    if list(labels).count(label) > 15:
        filtered_centers.append(cluster_centers[label])
        
print len(filtered_centers)
```

```python
with open("locations.txt", "a") as f:
    for center in filtered_centers:
        f.write(str(center[0]))
        f.write(',')
        f.write(str(center[1]))
        f.write('\n')
```

```python
def distance(l1, l2):
    return (l1[0] - l2[0])**2 + (l1[1] - l2[1])**2
```

```python
min_dists = []
for of in offices:
    dist = [(distance(of, x), x) for x in filtered_centers]
    print dist[:3]
    min_dists.append(min(dist))
    
print min(min_dists)
```

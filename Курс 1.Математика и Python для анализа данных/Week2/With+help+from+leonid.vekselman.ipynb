```python
import numpy as np
import re
from __future__ import print_function
f = open('sentences.txt', 'r')
lines = []
for l in f:
    l = l.lower()
    l = re.split('[^a-z]',l)
    l = filter(None, l)
    lines.append(l)

d = {}
# ну если на пальцах (без цикла)
print ('lines[0][0]=', lines[0][0])
print ('lines[0][1]=', lines[0][1])
print ('lines[0][2]=', lines[0][2])
# ...
print ('lines[1][0]=', lines[1][0])

d[0] = lines[0][0]
d[1] = lines[0][1]
d[2] = lines[0][2]
#...
d[11] = lines[1][0]
print ('d[0] =', d[0])
print ('d[1] =', d[1])
print ('d[2] =', d[2])
print ('d[11] =', d[11])

# вот таким "изящным" способом мне удалось записать в словарь данные в виде ключ: слово :-)
# для цикла, наверное, нужно сначало пробежать по словам первого предложения, потом переключиться на втрое предложение
# для этого нужен цикл в цикле, если я правильно понимаю ;-)
# пробую организовать новый словарь

word_index = dict()
i = 0
for sentence in lines:
    for word in sentence:
        if word not in word_index:
            word_index[word] = i
            i = i + 1
print (word_index)
# получилось! :-) но ключами выступают слова, а не цифры :-( Как сделать наоборот?
```
